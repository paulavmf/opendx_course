```python
import pandas as pd

# Read the CSV files into separate DataFrames
df1 = pd.read_csv('org_unit_can.csv')
df2 = pd.read_csv('org_units_cv.csv')
df3 = pd.read_csv('org_units_mauritania.csv')
df4 = pd.read_csv('org_units_sen.csv')

# Merge the DataFrames based on a common column
alldf = df1.append(df2).append(df3).append(df4)
# Save the merged DataFrame to a new CSV file
alldf.to_csv('merged_file.csv', index=False)
```
